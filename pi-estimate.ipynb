```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def monte_pi(trials):
    estimate = 0
    inside = 0
    x = np.random.uniform(0, 1, trials)
    y = np.random.uniform(0, 1, trials)
    inside = sum(x**2 + y**2 <= 1)
    inside_points = x**2 + y**2 <= 1
    estimate = (inside / trials) * 4
    plt.figure(figsize=(8, 8))
    plt.scatter(x[inside_points], y[inside_points], color='blue', s=1, label='Inside Circle')
    plt.scatter(x[~inside_points], y[~inside_points], color='red', s=1, label='Outside Circle')
    plt.title(f'Monte Carlo Estimation of Pi with {trials} Trials\nEstimated Pi = {estimate:.6f}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.axis('equal')
    plt.show()
    return estimate
```

```python
monte_pi(100)
```

```python
monte_pi(10000)
```

```python
monte_pi(500000)
```
