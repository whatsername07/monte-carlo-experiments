```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def monte_pi(trials):
    estimate = 0
    inside = 0
    for i in range(trials):
        x = random.random()
        y = random.random()
        if x**2 + y**2 <= 1:
            inside += 1
    estimate = (inside / trials) * 4
    return estimate

```

```python
monte_pi(10000000)
```
